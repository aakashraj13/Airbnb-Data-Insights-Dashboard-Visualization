## 3. ***Data Wrangling***

```python
# Handling missing values
df1.drop(columns=['neighbourhood_group'], inplace=True)
df2.drop(columns=['neighbourhood_group'], inplace=True)  


df1['price'].fillna(df1['price'].median(), inplace=True)
df2['price'].fillna(df2['price'].median(), inplace=True) 

df1['last_review'].fillna(pd.NaT, inplace=True)
df2['last_review'].fillna(pd.NaT, inplace=True) 

df1['reviews_per_month'].fillna(0, inplace=True)
df2['reviews_per_month'].fillna(0, inplace=True)

df1['host_name'].fillna("Unknown", inplace=True)
df2['host_name'].fillna("Unknown", inplace=True)

df1['license'].fillna("Unknown", inplace=True)
df2['license'].fillna("Unknown", inplace=True)

# Convert data types
df1['last_review'] = pd.to_datetime(df1['last_review'])
df2['last_review'] = pd.to_datetime(df2['last_review'])

df1['price'] = df1['price'].astype(int)
df2['price'] = df2['price'].astype(int)


# Removing outliers
price_cap1 = df1['price'].quantile(0.99)
df1 = df1[df1['price'] <= price_cap1]

price_cap2 = df2['price'].quantile(0.99)
df2 = df2[df2['price'] <= price_cap2]

min_nights_cap1 = df1['minimum_nights'].quantile(0.99)
df1 = df1[df1['minimum_nights'] <= min_nights_cap1]

min_nights_cap2 = df2['minimum_nights'].quantile(0.99)
df2 = df2[df2['minimum_nights'] <= min_nights_cap2]

# Standardizing column names
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")
```

```python
# Adding a new column to distinguish the datasets before merging
df1['city'] = 'Chicago'
df2['city'] = 'New Orleans'

# Merging the datasets
df = pd.concat([df1, df2], ignore_index=True)

# Display basic info after merging
print(df.info())

# Checking for missing values in the combined dataset
print(df.isnull().sum())
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1 Correlation Matrix

```python
# Select only numerical columns for correlation analysis
numeric_df = df.select_dtypes(include=['number'])

# Correlation matrix for numerical columns
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

##### 1. Why did you pick the specific chart?
I chose a heatmap because it visually represents correlations between numerical variables, making patterns easy to interpret. The color-coded scheme highlights strong positive and negative relationships, helping to identify multicollinearity and redundant features. Unlike scatter plots, a heatmap efficiently displays multiple correlations at once, aiding in feature selection and data-driven decision-making.

##### 2. What is/are the insight(s) found from the chart? 
The heatmap reveals strong correlations between certain numerical variables, helping identify key relationships. For example, if price is highly correlated with number of bedrooms or property size, it indicates that these factors significantly influence pricing. Additionally, weak or negative correlations highlight independent variables. This insight helps in feature selection for predictive modeling and understanding key drivers in the dataset.

#### Chart - 2 Price Distribution

```python
# Price Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['price'].dropna(), bins=50, kde=True)
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Price Distribution')
plt.show()
```

##### 1. Why did you pick the specific chart? 

A histogram with a KDE curve is ideal for visualizing the distribution of price values. It helps identify skewness, outliers, and common price ranges, making it useful for understanding pricing patterns and potential anomalies.

##### 2. What is/are the insight(s) found from the chart? 

The chart likely reveals whether price is normally distributed or skewed, indicating if most listings are within a certain price range or if there are extreme values (luxury listings or underpriced properties). If the distribution is right-skewed, it suggests a few high-price listings significantly impact the average price.
